# Análise dos metadados

```python
import pandas as pd
```

## Valores de precipitação maiores que zero

```python
df_test = pd.read_csv("../data/test_meta.csv", parse_dates=['dtInit', 'dtEnd'])
df_test.head()
```

```python
df_group = (df_test.groupby(pd.Grouper(key='dtInit', axis=0, freq='D'))
           .mean(numeric_only=True)
           .sort_values(by=['vlNonZero'], ascending=False)
           .reset_index()
           .head(10))

day = df_group['dtInit'].iloc[0].day
print(f"Dia com a maior média de valores de precipitção diferentes de zero: {day}")

df_group.head()
```

```python
df_day = df_test[df_test['dtInit'].dt.day==day]

idx = df_day.index[0]
print(f"Índice da primeira sequência do dia {day}: {idx}")

df_day.head()
```

## Média de precipitação

```python
df_test = pd.read_csv("../data/test_meta.csv", parse_dates=['dtInit', 'dtEnd'])
df_test.head()
```

```python
df_group = (df_test.groupby(pd.Grouper(key='dtInit', axis=0, freq='D'))
           .mean(numeric_only=True)
           .sort_values(by=['vlMean'], ascending=False)
           .reset_index()
           .head(10))

day = df_group['dtInit'].iloc[0].day
print(f"Dia com a maior média da média dos valores de precipitação: {day}")

df_group.head()
```

```python
df_day = df_test[df_test['dtInit'].dt.day==day]

idx = df_day.index[0]
print(f"Índice da primeira sequência do dia {day}: {idx}")

df_day.head()
```
